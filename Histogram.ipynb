# Histogram İşlemleri

Görüntünün piksellerinin grafiksel olarak gösteriminde sıklıkla kullanılan bir yöntemdir.Görüntü histogramı görüntünün tüm piksellerinin değerini tespit ederek piksel sayısını grafiksel olarak gösterir ancak piksellerin koordinatlarını göstermez.
Dikey piksel sayısı,yatay piksel değerlerini gösterir.


```
import PIL
from pylab import * 
goruntu = array(PIL.Image.open("barça.JPG"))
figure(1)
hist(goruntu.ravel(),bins = 255)

figure(2)
hist(goruntu.ravel(),bins = 64)

figure(3)
hist(goruntu.ravel(),bins = 8)
```

#Histogram Eşitleme

Bir görüntüdeki renk dağılımını düzenleyerek kontrast dengesini iyileştirmede histogram eşitleme yöntemi kullanılmaktadır.Global ve yerel olmak üzere ikiye ayrılır.

```
import cv2 as cv
import matplotlib.pyplot as plt 
import numpy as np 

renkli_img = cv.imread("barça.JPG")
gri_img = cv.imread("barça.JPG",0)
renkler = ["red","green","blue","gray"]

for i,v in enumerate(renkler):
  if i<3:
    histogram = cv.calcHist([renkli_img],[i],None,[256],[0,256])
  if i ==3 :
    histogram = cv.calcHist([gri_img],[0],None,[256],[0,256])
  plt.plot(histogram,color = v)
  plt.xlim([0,256])
plt.show()
```

```
import cv2 as cv
import matplotlib.pyplot as plt 
import numpy as np 
goruntu = cv.imread("14.png",0)
equ = cv.equalizeHist(goruntu)
ax = plt.subplot(211)
plt.hist(goruntu.flatten(),256,[0,256],color = "r")
ax1 = plt.subplot(212)
ax1.hist(equ.flatten(),256,[0,256],color = "b")
plt.show()
```

```
from google.colab.patches import cv2_imshow
res = np.hstack((goruntu,equ)) #We have stacked both images side-by-side.İki resmi de yan yana yığdık.Resimdeki ışıklandırılması eşitlendi.
cv2_imshow(res)
```
